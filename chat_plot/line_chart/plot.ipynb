```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import math
import numpy as np


params = {
    'legend.fontsize': 13,
    'legend.handlelength': 2,
    'legend.loc': 'lower right',
    'legend.markerscale': 1.5,
    'legend.edgecolor': 'black',
    'legend.borderaxespad': 0,
    'font.size': 13,
    'font.family': 'DeJavu Serif',
    'font.serif': ['Times New Roman'],
    'xtick.labelsize':'x-small',
    'ytick.labelsize':'x-small'
}
plt.rcParams.update(params)

# Assuming x, y1, and y2 are lists containing your data
x = np.array([0.0001, 0.001, 0.01, 0.1, 1])   # Number of temperature term
y1 = [1.12, 1.02, 0, 8.93, 9.46]  # Decrease of Recall@40(%) for line 1
y2 = [-0.82, -0.58,0, 0.26, 4.45]  # Decrease of Recall@40(%) for line 2
y = y1 + y2

fig, ax = plt.subplots()

fig.set_size_inches(4,3)

ax.plot(x.astype('str'), y1, color='red', label='LastFM', marker='v', mfc='none')
ax.plot(x.astype('str'), y2, color='green', label='MovieLens', marker='s', mfc='none')


# plt.title('title',**csfont)
# plt.xlabel('xlabel', **hfont)

plt.xlabel('$L_2$ regularization term $\\lambda_1$')
plt.ylabel('Decrease of NDCG@40 (%)')

new_list = np.arange(math.floor(min(y)-5), math.ceil(max(y)) + 2  , 2)
plt.yticks(new_list)
print(new_list)

# ax.xaxis.set_ticks(range(len(x)))
# ax.xaxis.set_ticklabels(x)
ax.xaxis.set_ticks(np.arange(len(x))) #set the ticks to be a
ax.xaxis.set_ticklabels(x) # change the ticks' names to x


ax.grid(True)  # Add this line to include gridlines
ax.grid(which='major',  linewidth=0.8, alpha=0.5)
ax.grid(which='minor',  linewidth=0.1)

plt.legend()
plt.show()
```

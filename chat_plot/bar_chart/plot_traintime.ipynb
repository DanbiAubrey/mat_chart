```python
import numpy as np
import matplotlib.pyplot as plt
import math

train_time_1 = [np.log(t* 1000) for t in [19.07 * 200, 32.27 * 200, 26.68 * 200 , 77.34 * 200 , 64.93* 200]]
test_time_1 = [np.log(t* 1000) for t in [3.61, 2.96, 1.66, 9.86, 0.58]]
train_time_2 = [np.log(t* 1000) for t in [33.05 * 200, 66.35 * 200 , 75.59 * 200, 129.74 * 200, 81.31 * 200]]
test_time_2 = [np.log(t* 1000) for t in [3.35, 2.74, 1.43, 8.95, 0.56]]
n = 5
r = np.arange(n)
width = 0.25

fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(10,3.5)

# Add labels for the legend
train_bar = ax1.bar(r, train_time_1, color='black', width=width, edgecolor='black', label='Training time')
test_bar = ax1.bar(r + width, test_time_1, color='white', width=width, edgecolor='black', label='Testing time')

ax2.bar(r, train_time_2, color='black', width=width, edgecolor='black')
ax2.bar(r + width, test_time_2, color='white', width=width, edgecolor='black')

ax1.set_xlabel("LastFM")
ax1.set_ylabel("Time(log(ms))")
ax2.set_xlabel("ML-1M")
ax2.set_ylabel("Time(log(ms))")

new_list = np.arange(0, math.ceil(max(train_time_1)), 2)
ax1.set_yticks(new_list)
ax1.set_xticks(r + width/2,['KGIN','KGCL','KGRec','KGAT', 'KHGRec'])

new_list = np.arange(0, math.ceil(max(train_time_1)), 2)
ax2.set_yticks(new_list)
ax2.set_xticks(r + width/2,['KGIN','KGCL','KGRec','KGAT', 'KHGRec'])

# Add grid
ax1.grid(axis='y', linewidth=0.8, alpha=0.3)
ax2.grid(axis='y', linewidth=0.8, alpha=0.3)

# Add legend to both subplots
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, ncols=2, loc='upper center')
plt.show()
```
